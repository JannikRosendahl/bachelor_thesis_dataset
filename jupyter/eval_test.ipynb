```python
# list of (executable, precision, recall, f1, support)
data = []

for line in result.split('\n'):
    x = line.split(' ')
    x = list(filter(lambda x: x != '', x))
    assert len(x) == 5
    # replace all nan with 0
    for i in range(1, 4):
        if x[i] == 'nan':
            x[i] = '0'
    data.append((x[0], float(x[1]), float(x[2]), float(x[3]), int(x[4])))
    
def print_data(data):
    # print first element left aligned padded to 20
    print('{:<20}'.format(data[0]), end='')
    # print rest right aligned padded to 10
    for d in data[1:]:
        print('{:>10.2f}'.format(d), end='')
    print()
    
```

```python
# calculate weighted and macro average
weighted = [0, 0, 0, 0]
macro = [0, 0, 0, 0]
total = 0
for d in data:
    total += d[4]
    for i in range(4):
        weighted[i] += d[4] * d[i+1]
        macro[i] += d[i+1]
for i in range(4):
    weighted[i] /= total
    macro[i] /= len(data)

print('Weighted average:', weighted)
print('Macro average:', macro)
```
