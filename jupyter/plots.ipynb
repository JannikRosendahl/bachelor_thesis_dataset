```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot - y_i * log(y_hat_i)

y_hat = np.linspace(0.01, 0.99, 100)
y = 1

plt.plot(y_hat, -y * np.log(y_hat))
plt.xlabel('y_hat')
plt.ylabel('crossentropy loss')
# use less space
plt.tight_layout()
# bigger font
plt.rcParams.update({'font.size': 15})

# save to svg
plt.savefig('ce_loss.svg')
plt.show()
```

```python
# plot - alpha_i * (1-y_hat_i) ** gamma * y_i * log(y_hat_i)

alpha = [1, 5, 10]
gamma = [2, 5]
y_hat = np.linspace(0.01, 0.99, 100)
y = 1

for g in gamma:
    for a in alpha:
        plt.plot(y_hat, -a * (1-y_hat) ** g * y * np.log(y_hat))
    #plt.plot(y_hat, - a * (1-y_hat) ** gamma * y * np.log(y_hat))
# use less space
plt.tight_layout()
# bigger font
plt.rcParams.update({'font.size': 15})

plt.xlabel('y_hat')
plt.ylabel('focal loss')
# legend
plt.legend([f'gamma={g}, alpha={a}' for g in gamma for a in alpha])


plt.savefig('focal_loss.svg')
plt.show()
```
